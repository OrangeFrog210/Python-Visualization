```python
"""Use NumPy to generate random data
"""
import numpy as np

# We'll use numpy to create data
x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
print(np.array([x, y]).reshape(25, 2))
print(np.array([x, y]).reshape(25, 2).reshape(2,25))
```

```python
"""It takes only 1 command to draw
"""
pylab.plot(x, y, 'b') # 'r' stands for red
# Tip: During presentations, keep colours consistent (e.g. if the company is using green as profit, mirror that in your plot)
```

```python
"""Drawing a subgraph
"""
# pylab gives the canvas
# The contents of the brackets represent (rows, columns, indices)

pylab.subplot(1,2,1) # plot column in row 1, column 2
pylab.plot(x,y,'r--') # pyplot (.plot) tells us what data
pylab.subplot(1,2,2)
pylab.plot(y,x,'b--')
```

```python
# in matplotlib, you can set other properties usch as line colour, transparency, and more
# transparency is especially useful if there are multiple data that crosses over
fig, axes = plt.subplots(dpi=150)
axes.plot(x,x**2, color='red', alpha=.5) # alpha=.5 means halfway seethrough
axes.plot(x,x+2, color='#1155dd', alpha=0.9)
axes.plot(x,x+3, color='black', alpha=0.3)
axes.plot(x,x+3, color='green', alpha=0.)
```

```python
fig, axes = plt.subplots(dpi=150)

# line width(= lw)
axes.plot(x,x+1, color='blue', linewidth=0.25)
axes.plot(x,x+2, color='blue', linewidth=1)
axes.plot(x,x+3, color='blue', linewidth=1.5)
axes.plot(x,x+4, color='blue', linewidth=2)
```

```python
fig, axes = plt.subplots(dpi=150)

# line style
axes.set_title('line style')
axes.plot(x,x+1, color='blue', lw=2, linestyle='-')
axes.plot(x,x+2, color='blue', lw=2, linestyle='-.')
axes.plot(x,x+3, color='blue', lw=2, linestyle=':')

# set dashes
line, = axes.plot(x,x+4,color='black',lw=1.50)
line.set_dashes([5,10,15,10])
```

```python
fig, axes = plt.subplots(dpi=100)

# marker
axes.set_title('Marker and associated parameters')
axes.plot(x,x+1, color='blue', lw=2, marker='o') # dots is the most common
axes.plot(x,x+2, color='blue', lw=2, marker='+')
axes.plot(x,x+3, color='blue', lw=2, marker='s', markersize=18, markerfacecolor='green') # square
axes.plot(x,x+4, color='blue', lw=2, marker='1')

# marker size
axes.plot(x,x-5, color='blue', lw=2, marker='1', markersize=5)

# marker colour
axes.plot(x,x+5, color='blue', lw=2, marker='o', markerfacecolor='red')
```

## Other 2D Graphics

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1,5,figsize=(16,5))
axes[0].set_title("Scatter")
axes[0].scatter(x,x+.25*np.random.rand(len(x)))
axes[1].set_title("Step")
axes[1].step(n,n**2,lw=2)
axes[1].set_title("Bar")
axes[1].step(n,n**2,lw=2) # two axes[1], becomes orange
axes[1].set_title("Bar")
axes[1].step(n,n**2,lw=2) # three axes[1], becomes green
axes[2].set_title("Bar")
axes[2].bar(n,n**2,align="center", alpha=.5,lw=2)
axes[3].set_title("Fill between", color='yellow')
axes[3].fill_between(n,n**2, color='black', alpha=.47)

axes[4].scatter(x,x+8*np.random.rand(len(x)))
```

```python
"""Draw a radar chart
"""
fig = plt.figure(figsize=(6,6))
# ax = fig.add_axes([0.0,0.0,.6,.6])
ax = fig.add_axes([0.0,0.0,.6,.6], polar=True)
t = np.linspace(0,2*np.pi,100)
ax.plot(t,t,color='brown', lw=10)
ax.plot(t,t*2,color='white', lw=10)
ax.plot(t,t*t,color='yellow', lw=4.5)
```

```python
"""Draw a histogram
"""
n = np.random.randn(100000)
fig, axes = plt.subplots(1,5,figsize=(12,4))
axes[0].set_title("Default histogram")
axes[0].hist(n)
axes[1].set_title("Default histogram")
axes[1].hist(n, cumulative=True, bins=10)
axes[2].set_title("Default histogram")
axes[2].hist(n, cumulative=True, bins=50)
axes[3].hist(n//2, color='red')
```

```python
"""Draw a histogram
"""
n = np.random.randn(100000)
fig, axes = plt.subplots(2,5,figsize=(12,4))
axes[0,0].set_title("Default histogram")
axes[0,0].hist(n)
axes[0,1].set_title("Default histogram")
axes[0,1].hist(n, cumulative=True, bins=10)
axes[1,2].set_title("Default histogram")
axes[1,2].hist(n, cumulative=True, bins=50)
axes[0,3].hist(n//2, color='red')
```

```python
# Draw a contour map
import matplotlib
import numpy as np
import matplotlib.cm as cm # colour map
import matplotlib.pyplot as plt # ply = pyplot

delta = .025
x = np.arange(-3.0, 3.0, delta) # create an array between -3 and 3, increments of delta
y = np.arange(-2.0,2.0,delta)
X, Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2) # exponential
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1 - Z2) * 2

print(X)
print(Y)
print(Z)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z) # create a conatour surface
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title("Contour map")

# MT: Can we create a map using contour map?
```
